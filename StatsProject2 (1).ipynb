```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the dataset
df = pd.read_excel("sample-superstore-subset-excel.xlsx")


# Display the first few rows and summary statistics
df.head(10)
```

```python

# Handle missing values
print(df.isnull().sum())

 # there are two ways for handling missing values, 1. imputation and 2. deletion

  # 1. imputaion - It involves filling the missing data with substituted values like mean, median or mode      

df['Product Base Margin'] = df['Product Base Margin'].fillna(df['Product Base Margin'].median())

  #2. deletion - It involves removing missing data entirely.
     
# Convert data types if necessary
 # Converting data types before analysis is crucial 
 # to ensure accurate calculations, prevent errors, optimize performance.
 # enable proper data handling and visualization. 

df['Order Priority'] = df['Order Priority'].astype('category')
df['Customer Name'] = df['Customer Name'].astype('category')
df['Ship Mode'] = df['Ship Mode'].astype('category')
df['Customer Segment'] = df['Customer Segment'].astype('category')
df['Product Category'] = df['Product Category'].astype('category')
df['Product Sub-Category'] = df['Product Sub-Category'].astype('category')
df['Product Container'] = df['Product Container'].astype('category')
df['Product Name'] = df['Product Name'].astype('category')
df['Region'] = df['Region'].astype('category')
df['State or Province'] = df['State or Province'].astype('category')
df['City'] = df['City'].astype('category')

print(df.dtypes)

# Create any new columns needed for analysis

```

```python

# Univariate analysis
 # It is a way to look at and describe data for just one thing at a time.
 # It helps you understand how that single variable behaves, including where most of the data is centered.
 # how spread out the data is, and what its overall distribution looks like.
 # It doesnâ€™t look at how this variable relates to others, just its own characteristics.

df.info()
df.describe(include = "all")

 # histogram for sales 

plt.figure(figsize=(6, 4))
sns.histplot(df["Sales"], kde=True)
plt.title("Distribution of sales")
plt.xlabel("Sales")
plt.ylabel('count')
plt.show()

 # barplot for Discount

sns.barplot(x='Region', y='Sales', data=df )
plt.title('Count of Sales by Region')
plt.ylabel("Sales")
plt.xlabel('Region')
plt.show()



# Bivariate analysis
 # It is a way to look at how two variables are connected.
 # it checks how two variables interact, how they might affect each other, or if they are related in some way.

 # Scatter plot for  Sales vs Profit
plt.figure(figsize=(8, 6))
plt.scatter(x=df["Sales"], y=df["Profit"])
plt.title('Sales vs Profit')
plt.xlabel("Sales")
plt.ylabel('Profit')
plt.show()

 # Box plot for Region vs Sales
sns.boxplot(x='Region', y='Sales', data=df)
plt.title('Region vs Sales')
plt.ylabel("Sales")
plt.xlabel('Region')
plt.show()




```

## Additional Statistical Analysis


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import scipy.stats as stats
from sklearn.preprocessing import LabelEncoder
import copy
from scipy.stats import ttest_ind, chi2_contingency

# Example: T-test for sales in different regions

HO = "sales for Region" 
Ha = "sales for no Region"


x = np.array(df[df.Region == "yes"]['Sales'])
y = np.array(df[df.Region == "no"]["Sales"])


t, p_value = stats.ttest_ind(x,y, axis = 0)
print(f'T-statistic: {t:.4f}')
print(f'P-value: {p_value:.4f}')


if p_value < 0.05:
    print(f'{Ha} as the p_value ({p_value}) < 0.05')
else:
    print(f'{HO} as the p_value ({p_value}) > 0.05')



# Example: Chi-square test for independence between Ship Mode and Customer Segment

contingency_table = pd.crosstab(df['Ship Mode'], df['Customer Segment'])

chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

print(f'\nChi-square Statistic: {chi2_stat:.4f}')
print(f'P-value: {p_value:.4f}')
print(f'Degrees of Freedom: {dof}')

alpha = 0.05
if p_value < alpha:
    print("\nThere is a significant association between Ship Mode and Customer Segment.")
else:
    print("\nThere is no significant association between Ship Mode and Customer Segment.")

```

```python

# Plotting sales trends over time
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

plt.figure(figsize=(10, 6))
plt.plot(df['Ship Date'], df['Sales'], marker='o', linestyle='-', color='b')

plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)

# Show the plot
plt.show()

# Resample sales data by month

df['Ship Date'] = pd.to_numeric(df['Ship Date'], errors='coerce')

plt.plot(df['Ship Date'], df['Sales'])
plt.title('Sales Trends Over month')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python

# Plotting profit trends over time
df['Order Date'] = pd.to_datetime(df['Order Date'])

df.set_index('Order Date', inplace=True)

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Profit'],color='green')

plt.title('Profit Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Profit')
plt.grid(True)

plt.show()





```

```python

# Identifying top-performing product categories based on total sales and profit

category_performance = df.groupby('Product Category').agg({'Sales': 'sum', 'Profit': 'sum'})

top_categories_sales = category_performance.sort_values(by='Sales', ascending=False)

top_categories_profit = category_performance.sort_values(by='Profit', ascending=False)

print("Top-Performing Product Categories by Sales:")
print(top_categories_sales)

print("\nTop-Performing Product Categories by Profit:")
print(top_categories_profit)

top_categories_sales.plot(kind='bar', figsize=(10, 6), title='Top-Performing Product Categories by Sales')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()


top_categories_profit.plot(kind='bar', figsize=(10, 6), title='Top-Performing Product Categories by Profit', color=['orange', 'green'])
plt.xlabel('Product Category')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()


# Plotting total sales by product category

total_sales_by_category = df.groupby('Product Category')['Sales'].sum()

plt.figure(figsize=(10, 6))
total_sales_by_category.plot(kind='bar', color='skyblue')

plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()





# Plotting total profit by product category

total_profit_by_category = df.groupby('Product Category')['Profit'].sum()

plt.figure(figsize=(10, 6))
total_profit_by_category.plot(kind='bar', color='lightcoral')

plt.title('Total Profit by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()





```
